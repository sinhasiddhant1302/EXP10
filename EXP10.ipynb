```python
#Create DataFrame
data={                              #stored in dictionary
    "Name":["A","B","C"],
    "Marks":[85,90,78]
}
df= pd.DataFrame(data)              # used to convert data ito tabular form
print(df)
```

```python
print(df.head())                           #prints first 5 rows
print(df.tail())                           #prints last 5 rows
print(df.head(2))                          #prints first 2 rows
print(df.tail(2))                          #prints last 2 rows
```

```python
# Find the structure of the DataFrame
print(df.shape)                         #gives you the no. of rows and colums
```

```python
#Dimensions
print(df.ndim)                          # 2 dimension as it has only name and marks i.e total no. of variables/parameters.
```

```python
print(df.size)                         #total number of elements
```

```python
#Display Column Names
print(df.columns)
```

```python
#Check Data Type of each Column
df.dtypes
```

```python
#Access Individual Column
print(df["Name"])
df["Name"]
display(df["Name"])
```

```python
#Access specific row
df.loc[1]                       #prints 2nd row as it has indexing 1 so it prints a specific row
```

```python
# loc and iloc are different as to print 1st row 2nd column type of things can be printed by iloc
df.iloc[1,1]
```

```python
#Add a New Column
'''If marks 80-90- First Class
If marks>=90- Distinction
If marks >80- Second Class'''
df["Grade"]=["First class","Distinction","Second Class"]          #used to add a new column and adds it in the sequence you've typed it in.
df                                                                #prints df
```

```python
#Update data in DataFrame
#Change marks of student A to 88.
df.loc[0,"Marks"]= 81
df
df.iloc[0,1]=81
df
```

```python
#Delete a Column
df1=df.drop("Grade",axis=1)
display(df1)
display(df)
```

```python
#Perform Basic Statistical Analysis
mean1=df["Marks"].mean()
mean1
```

```python
max1=df["Marks"].max()
print(max1)
min1=df["Marks"].min()
print(min1)
```

```python
#Apply Condition(filtering)
#Display students scoring more than 80
df[df["Marks"] > 80]
```
